```python
cars['model_year'].fillna('Missing', inplace = True)
cars['cylinders'].fillna(0, inplace = True)
cars['paint_color'].fillna('Missing', inplace = True)
cars['is_4wd'].fillna(0, inplace = True)
cars['odometer'].fillna(0, inplace = True)
```

The na values of 0 were changed to floats from string values by instruction of reveiwer. All string nan values changed to 'missing'. 

```python
cars.head(10)
```

```python
grouped_data = cars.groupby(['model', 'model_year'])['cylinders']
grouped_data.head(5)
```

Groupby is used here as requested by reviewer. The model of the cars and year of car go up as cylinder count go up. 

```python
grouped_data.median()
```

Median od these models by year and model type were calculated.

```python
cylinder_nan_fillna = cars.groupby(["model_year"])['cylinders'].median().reset_index()
```

```python
cylinder_nan_fillna
```

```python
def replace_with_median(row):
    if pd.isna(row['cylinders']):
        return cylinder_nan_fillna[cylinder_nan_fillna['model_year']] == row['model_year']
    return row['cylinders']

cars['cylinders_upd'] = cars.apply(replace_with_median, axis = 1)
```

This variable and function was made to fix na values of cylinders by request of reviewer. Function replaces nan values with median cylinder values based on 
model year and model type.

```python
cars['cylinders'].isna().sum()
```

All nan values have been fixed.

```python
fig = px.histogram(x = cars['model_year'], y = cars['price'])
fig.update_layout(xaxis_title='Model Year', yaxis_title='Price') #app.py style for convinence
fig.update_layout(title = 'Car Sales Histogram')
fig.show()
```

As shown in this grapgh, cars tend to be bought at higher rates with higher prices the more recent the model. The deep dive in price can be explained by bonus discounts and overall bonuses given to people for trading in there older model to get newer models. Essentially the either people dont buy the latest cars because they maybe too expensive, or they dont have to buy them full price because they pay a trade in discount for the vechicle.
